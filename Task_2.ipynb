```python
dfs = pd.read_excel("School_round_exhibit-Prerit.xlsx",sheet_name=None)
df = dfs['Data Exhibit-2']
df.head()
```

```python
# for each sku, do the following simulation 
samp_df = df.iloc[0]
```

```python
inv_list = []
order_list = []

for idx in range(len(df)):
    avg_inv = 0
    avg_order = 0
    samp_df = df.iloc[idx]
    for _ in range(5):
        opening_inv = samp_df['Opening Inventory (units)']
        lead_time = samp_df['Revised Lead Time']
        mu = samp_df['Demand During Lead time']
        sigma = samp_df['Sigma during Lead Time']
        allowed_inv = samp_df['95% SL']
        order_place = False
        order_pending = False
        curr_inv = opening_inv
        day_since_ord_p = 0
        num_order_placed = 0
        # for daily demand - do it for 365 days
        for i in range(0,366):
            if order_place:
                day_since_ord_p += 1
        #         print("Days since order placed : {}".format(day_since_ord_p))
                if day_since_ord_p == lead_time:
                    order_pending = False
                    order_place = False
                    curr_inv += allowed_inv
                    day_since_ord_p = 0
                else:
                    order_pending = True
            demand = np.random.normal(mu, sigma, 1)[0]
            if demand <= 0:
                demand = 0
        #     print(curr_inv, demand)
            curr_inv = curr_inv-demand
            if curr_inv < 0:
                curr_inv = 0
            if curr_inv < allowed_inv:
                if (order_pending==False):
                    order_place = True
        #             print("Order Placed")
                    num_order_placed += 1
        
        avg_inv += curr_inv
        avg_order += num_order_placed
    
    inv_list.append(avg_inv/5.0)
    order_list.append(avg_order/5.0)
```
